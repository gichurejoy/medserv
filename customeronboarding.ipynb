```python
import pandas as pd
import numpy as np
import os


file_path = '/content/drive/MyDrive/Customer Biodata.xlsx'
customer_df = pd.read_excel(file_path)

print(customer_df.head())
```

```python
from re import template
file_path = '/content/drive/MyDrive/customeronboarding.xlsx'
template_df = pd.read_excel(file_path)

print(template_df.head())
```

```python
column_mapping = {
    'ACCOUNT': 'account',
    'CUSTOMER NAME': 'name',
    'REGION': 'country_id',
    'ZIP': 'zip',
    'ADDRESS':'street',
    'PHONE': 'phone',
    'TOWN': 'city',
    'EMAIL': 'email',

}
```

```python
customer_df.rename(columns=column_mapping, inplace=True)

# Add missing columns to df_supplier with NaN values
for col in template_df.columns:
    if col not in customer_df.columns:
        customer_df[col] = np.nan

# Set the is_company logic
customer_df['is_company'] = customer_df['company_name'].apply(
    lambda x: 0 if pd.notna(x) and x.strip().upper().startswith('DR') else 1 if pd.notna(x) and x.strip() != '' else 0
)
```

```python
# Set default values for country_id and state_id
customer_df['country_id'] = 'Kenya'
customer_df['state_id'] = 'KE'

customer_df = customer_df[template_df.columns]

# Drop the "Unnamed" columns past the 'bank_ids/acc_number' column
columns_to_keep = list(template_df.columns[:template_df.columns.get_loc('bank_ids/acc_number') + 1])
customer_df = customer_df[columns_to_keep]

# Save the new DataFrame to a CSV file
output_file = os.path.join('new_customer_template.csv')
customer_df.to_csv(output_file, index=False)

print(f"New customer template file saved to {output_file}")
```
